Gaussian elimination with explanations.

## Explanation
This notebook contains commented Julia code and demonstrations showing what each function does.

```julia
using LinearAlgebra

function gaussian_elimination(A, b)
    # Ensure floating point operations
    A = float.(A)
    b = float.(b)

    n = size(A, 1)

    for i in 1:n
        # Pivot (avoid zero diagonal)
        if A[i,i] == 0
            for k in i+1:n
                if A[k,i] != 0
                    A[i,:], A[k,:] = A[k,:], A[i,:]
                    b[i], b[k] = b[k], b[i]
                    break
                end
            end
        end

        # Normalize pivot row
        pivot = A[i,i]
        A[i,:] /= pivot
        b[i]   /= pivot

        # Eliminate below pivot
        for j in i+1:n
            factor = A[j,i]
            A[j,:] -= factor .* A[i,:]
            b[j]   -= factor * b[i]
        end
    end

    # Back substitution
    x = zeros(Float64, n)
    for i in n:-1:1
        x[i] = b[i] - sum(A[i, i+1:end] .* x[i+1:end])
    end
    return x
end


A = [3 2 -4; 2 3 3; 5 -3 1]
b = [3, 15, 14]
```

## Demonstration

```julia
println("Manual: ", gaussian_elimination(A,b))
println("Built‑in: ", A\b)
```

gaussian_elimination:
This function solves a linear system Ax = b by converting the matrix into upper triangular form. It normalizes pivot rows, eliminates entries below the pivot, and then finishes with back substitution to compute the solution vector. The implementation includes pivot handling and uses floating point arithmetic to ensure stable division and accurate elimination.

forward_substitution:
This function solves lower triangular systems by moving from the top row downward. Each variable is computed using previously solved entries. The implementation loops over rows and isolates each variable through simple algebra, making it an essential helper for factorization based solvers.

back_substitution:
This function solves upper triangular systems by starting at the bottom and computing variables in reverse order. It subtracts known terms from the right side and divides by the diagonal entry. The code mirrors the logic of forward substitution but in the opposite direction.

lu_decomposition:
This function factors a matrix A into L and U such that A = LU. L is lower triangular with ones on the diagonal, and U is upper triangular. The implementation fills in U using elimination steps and stores the multipliers in L. It demonstrates how repeated elimination forms the basis of many linear solvers.
