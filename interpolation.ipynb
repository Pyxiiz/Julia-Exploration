This notebook implements:
- Lagrange interpolation
- Newton divided differences
- Polynomial evaluation


```julia

using Plots

function lagrange_interp(x, y, x0)
    n = length(x)
    L = zeros(n)
    for i in 1:n
        L[i] = prod((x0 - x[j]) / (x[i] - x[j]) for j in 1:n if j != i)
    end
    return sum(y[i] * L[i] for i in 1:n)
end

x = [0, 1, 2, 3]
y = [1, 2, 0, 5]

xx = range(0, 3, length=200)
yy = [lagrange_interp(x, y, xi) for xi in xx]

plot(xx, yy, label="Interpolant")
scatter!(x, y, label="Data")
```
