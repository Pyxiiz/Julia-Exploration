# Linear Algebra Project — Julia

This notebook implements:
- Gaussian Elimination
- LU Factorization (manual)
- Solving linear systems


```julia

using LinearAlgebra

function gaussian_elimination(A, b)
    A = copy(A)
    b = copy(b)
    n = size(A, 1)

    for k in 1:n-1
        for i in k+1:n
            factor = A[i, k] / A[k, k]
            A[i, k:end] .-= factor .* A[k, k:end]
            b[i] -= factor * b[k]
        end
    end

    x = zeros(n)
    for i in n:-1:1
        x[i] = (b[i] - A[i, i+1:end] ⋅ x[i+1:end]) / A[i, i]
    end
    return x
end

A = [3 2 -4; 2 3 3; 5 -3 1]
b = [3, 15, 14]

x = gaussian_elimination(A, b)
println("Solution: ", x)
```
