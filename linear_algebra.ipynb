Gaussian elimination with explanations.

## Explanation
This notebook contains commented Julia code and demonstrations showing what each function does.

```julia
using LinearAlgebra

function gaussian_elimination(A, b)
    # Ensure floating point operations
    A = float.(A)
    b = float.(b)

    n = size(A, 1)

    for i in 1:n
        # Pivot (avoid zero diagonal)
        if A[i,i] == 0
            for k in i+1:n
                if A[k,i] != 0
                    A[i,:], A[k,:] = A[k,:], A[i,:]
                    b[i], b[k] = b[k], b[i]
                    break
                end
            end
        end

        # Normalize pivot row
        pivot = A[i,i]
        A[i,:] /= pivot
        b[i]   /= pivot

        # Eliminate below pivot
        for j in i+1:n
            factor = A[j,i]
            A[j,:] -= factor .* A[i,:]
            b[j]   -= factor * b[i]
        end
    end

    # Back substitution
    x = zeros(Float64, n)
    for i in n:-1:1
        x[i] = b[i] - sum(A[i, i+1:end] .* x[i+1:end])
    end
    return x
end


A = [3 2 -4; 2 3 3; 5 -3 1]
b = [3, 15, 14]
```

## Demonstration

```julia
println("Manual: ", gaussian_elimination(A,b))
println("Built‑in: ", A\b)
```
